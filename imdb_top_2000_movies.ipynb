```python
# An√°lise de Dados do Conjunto IMDb
# Alunos: Vitor Shimizu Farias de Campos 550390
# Daniel Barborsa Mariano 551171

import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('IMDB_Top_2000_Movies.csv')
```

```python
# Quest√£o 1: Distribui√ß√£o dos filmes por g√™nero
genre_counts = df['Genre'].value_counts()
genre_counts.plot(kind='bar', figsize=(10, 6), color='skyblue')
plt.title('Distribui√ß√£o dos filmes por g√™nero')
plt.xlabel('G√™nero')
plt.ylabel('N√∫mero de filmes')
plt.xticks(rotation=45)
plt.show()
```

```python
# Quest√£o 2: Distribui√ß√£o das avalia√ß√µes dos filmes
df['IMDB Rating'].plot(kind='hist', bins=20, figsize=(10, 6), color='salmon')
plt.title('Distribui√ß√£o das avalia√ß√µes dos filmes')
plt.xlabel('Avalia√ß√£o do IMDb')
plt.ylabel('N√∫mero de filmes')
plt.show()
```

```python
# Quest√£o 3: Rela√ß√£o entre a dura√ß√£o dos filmes e suas avalia√ß√µes
plt.scatter(df['Duration'], df['IMDB Rating'], alpha=0.5, color='green')
plt.title('Rela√ß√£o entre dura√ß√£o dos filmes e avalia√ß√µes')
plt.xlabel('Dura√ß√£o (minutos)')
plt.ylabel('Avalia√ß√£o do IMDb')
plt.show()
```

```python
# Quest√£o 4: Top 10 diretores com mais filmes no conjunto de dados
top_directors = df['Director'].value_counts().head(10)

plt.figure(figsize=(10, 6))
top_directors.plot(kind='bar', color='lightcoral')
plt.title('Top 10 diretores com mais filmes')
plt.xlabel('Diretor')
plt.ylabel('N√∫mero de filmes')
plt.xticks(rotation=45)
plt.show()
```

```python
# Quest√£o 5: Melhores avalia√ß√µes de cada g√™nero
best_ratings = df.groupby('Genre')['IMDB Rating'].max()
plt.figure(figsize=(10, 6))
best_ratings.plot(kind='bar', color='salmon')
plt.title('Melhores avalia√ß√µes de cada g√™nero')
plt.xlabel('G√™nero')
plt.ylabel('Avalia√ß√£o do IMDb')
plt.xticks(rotation=45)
plt.show()
```

```python
# Quest√£o 6: Popularidade dos filmes ao longo do tempo
df['Release Year'].value_counts().sort_index().plot(kind='line', figsize=(10, 6), color='purple')
plt.title('Popularidade dos filmes ao longo do tempo')
plt.xlabel('Ano')
plt.ylabel('N√∫mero de filmes')
plt.show()
```

```python
# Quest√£o 7: N√∫mero de g√™neros de filmes
num_genres = df['Genre'].str.split(',').explode().nunique()
genre_counts = df['Genre'].str.split(',').explode().value_counts()

plt.figure(figsize=(10, 6))
genre_counts.plot(kind='bar', color='skyblue')
plt.title('N√∫mero de filmes por g√™nero')
plt.xlabel('G√™nero')
plt.ylabel('N√∫mero de filmes')
plt.xticks(rotation=45)
plt.show()

print("N√∫mero de g√™neros de filmes:", num_genres)
```
